```python
import os

# Directories for wav and npy files
wav_dir = "recorded_audio"
npy_dir = "recorded_npy"

# Sort the files in each directory to ensure the order is consistent
wav_files = sorted([f for f in os.listdir(wav_dir) if f.endswith('.wav')])
npy_files = sorted([f for f in os.listdir(npy_dir) if f.endswith('.npy')])

# Rename .wav files to data_1.wav, data_2.wav, etc.
for i, filename in enumerate(wav_files, start=1):
    new_name = f"data_{i}.wav"
    old_path = os.path.join(wav_dir, filename)
    new_path = os.path.join(wav_dir, new_name)
    os.rename(old_path, new_path)
    print(f"Renamed {old_path} to {new_path}")

# Rename .npy files to data_1.npy, data_2.npy, etc.
for i, filename in enumerate(npy_files, start=1):
    new_name = f"data_{i}.npy"
    old_path = os.path.join(npy_dir, filename)
    new_path = os.path.join(npy_dir, new_name)
    os.rename(old_path, new_path)
    print(f"Renamed {old_path} to {new_path}")
```

```python
import os
from sklearn.model_selection import train_test_split

mel_dir = r'C:\Users\\danhearn\\Documents\\GitHub\\diffusertrack\\recorded_npy'
wav_dir = r'C:\Users\\danhearn\\Documents\\GitHub\\diffusertrack\\recorded_audio'
mel_files = [f for f in os.listdir(mel_dir) if f.endswith('.npy')]

# Create file paths with both .npy and .wav paths
# Create file paths with only filenames
paired_files = [
    f"{mel_file}|{mel_file.replace('.npy', '.wav')}"
    for mel_file in mel_files if os.path.exists(os.path.join(wav_dir, mel_file.replace('.npy', '.wav')))
]

# Check if paired_files has entries
if len(paired_files) == 0:
    print("No matching .npy and .wav file pairs found.")
else:
    # Split into training and validation sets
    train_files, val_files = train_test_split(paired_files, test_size=0.1, random_state=42)

    # Write only filenames to `training.txt` and `validation.txt`
    with open('training.txt', 'w') as f:
        for item in train_files:
            f.write("%s\n" % item)

    with open('validation.txt', 'w') as f:
        for item in val_files:
            f.write("%s\n" % item)

print("Training and validation files created with filenames only.")
```

```python
import os
from sklearn.model_selection import train_test_split

# Get the base filenames (without extensions) for pairs that exist in both directories
mel_files = [os.path.splitext(f)[0] for f in os.listdir(mel_dir) if f.endswith('.npy')]
wav_files = [os.path.splitext(f)[0] for f in os.listdir(wav_dir) if f.endswith('.wav')]
base_names = list(set(mel_files) & set(wav_files))  # Keep only names that have both .wav and .npy

# Split into training and validation sets
train_files, val_files = train_test_split(base_names, test_size=0.1, random_state=42)

# Write to `training.txt` and `validation.txt` with base names only
with open('training.txt', 'w') as f:
    for item in train_files:
        f.write("%s\n" % item)

with open('validation.txt', 'w') as f:
    for item in val_files:
        f.write("%s\n" % item)

print("Created `training.txt` and `validation.txt` with base names only.")
```

```python
import os
import librosa
import soundfile as sf

# Directory containing your .wav files
wav_dir = 'recorded_audio'
target_sr = 22050

# Resample each .wav file
for filename in os.listdir(wav_dir):
    if filename.endswith('.wav'):
        file_path = os.path.join(wav_dir, filename)
        y, sr = librosa.load(file_path, sr=target_sr)  # Load and resample
        sf.write(file_path, y, target_sr)  # Overwrite with resampled audio
        print(f"Resampled {filename} to {target_sr} Hz")
```
